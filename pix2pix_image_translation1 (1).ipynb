```python
# STEP 2: Import required libraries

import torch
# torch → main deep learning library (runs the AI model)

from diffusers import StableDiffusionInstructPix2PixPipeline
# StableDiffusionInstructPix2PixPipeline → special model for image-to-image editing

from PIL import Image
# Image → used to open and process images

import requests
# requests → used to download image from internet

from io import BytesIO
# BytesIO → converts downloaded data into image format
```

```python
pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
    "timbrooks/instruct-pix2pix",
    torch_dtype=torch.float32
)

pipe.enable_attention_slicing()
pipe = pipe.to("cpu")
```

```python
url = "https://images.unsplash.com/photo-1503023345310-bd7c1de61c7d"

response = requests.get(url)
image = Image.open(BytesIO(response.content)).convert("RGB")

image
```

```python
image = image.resize((128, 128))
```

```python
# STEP 5: Define instruction (prompt)

# This tells the model HOW to modify the image
# You can change this to anything creative

prompt = "a realistic oil painting of the scene"
"watercolor painting style"
"high quality artistic painting"
```

```python
# STEP 6: Generate edited image (SAFE VERSION)

image = image.resize((256, 256))  # increase size

edited_image = pipe(
    prompt=prompt,
    image=image,
    num_inference_steps=10   # increase steps
).images[0]

edited_image
```

```python
image = image.resize((256, 256))

edited_image = pipe(
    prompt="a high quality realistic oil painting of the scene",
    image=image,
    num_inference_steps=15
).images[0]

edited_image
```
